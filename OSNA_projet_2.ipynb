```python
%pip install python-igraph
%pip install leidenalg
import pandas as pd
from igraph import Graph
import leidenalg as la

from igraph import Graph
import leidenalg as la
from google.colab import drive
drive.mount('/content/drive')
# Read the CSV file containing the edge list
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/large_twitch_edges.csv')

# Create a graph from the edge list
G = Graph.TupleList(df.itertuples(index=False), directed=False)
```

```python
# Apply the Leiden algorithm to find communities
partition = la.find_partition(G, la.ModularityVertexPartition)

# Extract communities
communities = {i: [node for node in partition] for i, partition in enumerate(partition)}

# Print communities
for i, community in communities.items():
    print(f'Community {i+1}: {community}')
```

```python
total = 0
for i, community in communities.items():
    print(len(community))
    total += len(community)
print(total)
```

```python
# Define a function to find the node with maximum centrality measure in a community
def find_max_centrality_node(community, measure):
    if measure == "betweenness":
        centralities = community.betweenness(directed=False)
    elif measure == "degree":
        centralities = community.degree()
    elif measure == "eigenvector":
        centralities = community.eigenvector_centrality(directed=False)
    max_index = centralities.index(max(centralities))
    return community.vs[max_index]["name"]

# Iterate over each community and find the nodes with maximum centrality measures
for i, community in enumerate(partition.subgraphs()):
    if community.vcount() > 12000:
      continue
    print(f'Community {i+1}:')
    max_betweenness_node = find_max_centrality_node(community, "betweenness")
    max_degree_node = find_max_centrality_node(community, "degree")
    max_eigenvector_node = find_max_centrality_node(community, "eigenvector")
    print(f'Max Betweenness Node: {max_betweenness_node}')
    print(f'Max Degree Node: {max_degree_node}')
    print(f'Max Eigenvector Centrality Node: {max_eigenvector_node}')
    print()
```

```python
# Iterate over each community and find the nodes with maximum centrality measures
for i, community in enumerate(partition.subgraphs()):
    if community.vcount() > 12000:
      continue
    print(f'Community {i+1}:')
    nodes = community.vs["name"]

    # Filter edges that contain nodes from the last community
    filtered_edges = df[(df['numeric_id_1'].isin(nodes)) | (df['numeric_id_2'].isin(nodes))]

    # Export the filtered edges to a new CSV file
    filtered_edges.to_csv('community_'+str(i+1)+'.csv', index=False)
```

```python
from collections import defaultdict

# Create a dictionary to store the communities each node is present in
node_communities = defaultdict(list)

# Iterate over each community and populate the dictionary
for i, community in enumerate(partition.subgraphs()):
    if community.vcount() > 12000:
      continue
    communitycsv = pd.read_csv('community_'+str(i+1)+'.csv')
    for node in pd.concat([communitycsv['numeric_id_1'], communitycsv['numeric_id_2']]).unique():
        node_communities[node].append(i)

print(total)
# Find a node that is present in exactly two communities
for node, communities in node_communities.items():
    if len(communities) == 2:
        print(communities)
```
